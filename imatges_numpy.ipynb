```python
#Exercici 8
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import urllib
%matplotlib inline

img = mpimg.imread(
    urllib.request.urlopen("https://i.redd.it/cuvkibsq32651.jpg"),
    format="jpg")
print("La forma de la imatge en matriu és: ", img.shape)
print(plt.imshow(img))

#eliminem el verd
img_exverd = np.copy(img)
img_exverd[:,:,1]=0


plt.title("Ex verd")
plt.show(plt.imshow(img_exverd))

#eliminem el blau
img_exblau = np.copy(img)
img_exblau [:,:,2]=0

plt.title("Ex blau")
plt.show(plt.imshow(img_exblau))


mpimg.imsave ("Ex verd.jpg",img_exverd)
mpimg.imsave ("Ex blau.jpg",img_exblau)
```
